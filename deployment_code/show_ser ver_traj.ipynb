```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("traj.csv")
```

```python
# Make 3d plot of x,y,z
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(df["x"], df["y"], df["z"], label='3D Trajectory')
ax.plot(df["gx"], df["gy"], df["gz"], label='3D Goal')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(0,3)
plt.show()
```
